```python
# import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search for acetylcholinesterase
target = new_client.target
target_query = target.search('acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# select and retrieve bioactivity data for acetylcholinesterase on Homo sapien

selected_target = targets.target_chembl_id[0]
selected_target
```

```python
# report the ic50 in nanomolar for acetylcholinesterase
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type='IC50')
df = pd.DataFrame.from_dict(res)
df
```

```python
df.to_csv('acetylcholinesterase_bioactivity_data.csv', index=False)
```

```python
# handling missing data on standard value column and dropping them
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
len(df2.canonical_smiles.unique())
```

```python
#remove duplacate values on canonical smiles column
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

## Data pre-processing of bioactivity data

### Labeling compounds as either being active, inactive or intermediate

The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. As for those values in between 1,000 and 10,000 nM will be referred to as intermediate.

```python
selected_col = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selected_col]
df3
```

```python
df3.isnull().sum()
```

```python
bioactivity_class = []
for i in df3.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append('inactive')
  elif float(i) <= 1000:
    bioactivity_class.append('active')
  else:
    bioactivity_class.append('intermediate')
```

```python
len(bioactivity_class)
```

```python
df3.loc[:,'class'] = bioactivity_class
df3
```

## Save dataframe file to filepath

```python
df3.to_csv('acetylcholinesterase_bioactivity_preprocessed_data.csv', index=False)
```
